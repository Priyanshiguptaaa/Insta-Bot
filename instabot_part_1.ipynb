## 4 : Follow/Unfollow given handle - 

#### 1.Open the Instagram Handle of “So Delhi”
#### 2.Start following it. Print a message if you are already following
#### 3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def following(username):
    
    #searching the handle in the search box
    search=driver.find_element_by_xpath('//input[contains(@class,"XTCLo")]')  
    search.send_keys(username)
    
    #wait until the handle is visible
    link=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a')))  

    #opening the handle
    link.click()  
    
    #waiting for the follow button to appear
    follow=wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_5f5mN")]'))) 
    
    #checking if the account is being followed already
    if follow.text=='Follow': 
        
        print("not following")
        follow.click()
        print("followed now")
        
    else:
        print("following")  
```

```python
def unfollowing(username):
    search=driver.find_element_by_xpath('//input[contains(@class,"XTCLo")]')
    search.send_keys(username)
    
    link=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a')))

    link.click()
    follow=wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_5f5mN")]')))
    
    if follow.text!='Follow':
        print("following")
        follow.click()
        
    else:
        unfollow=driver.find_element_by_xpath('//button[contains(@class,"aOOlW")]')
        unfollow.click()
        print("not following")
```

```python
following('So Delhi')
```

## 5 : Like/Unlike posts

#### 1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
#### 2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def like_post(string):
    
    #opening the profile
    driver.get('https://www.instagram.com/' + string)  
    time.sleep(2)
    
    #opening the first post
    url=driver.find_element_by_class_name('_9AhH0')
    url.click()
    time.sleep(2)
                       
    #liking the post givn in the range
    for i in range(30): 
        
        #searching for the like button
        like=driver.find_element_by_xpath('//span/button[@class="wpO6b "]') 
        
        #clicking the like button              
        like.click()  
        time.sleep(1)
        next_pic=driver.find_element_by_xpath('//div[@class="DdSX2"]/a[contains(@class,"coreSpriteRightPaginationArrow")]')
        next_pic.click()
        time.sleep(2)                   
```

```python
like_post('dilsefoodie')
```

```python
def unlike_post(string):
    
    driver.get('https://www.instagram.com/' + string)
    time.sleep(2)
    img_url=driver.find_element_by_class_name('_9AhH0')
    img_url.click()
    time.sleep(2)
    for i in range(30):
        like=driver.find_element_by_xpath('//span/button[@class="wpO6b "]')
        like.click()
        time.sleep(1)
        next_pic=driver.find_element_by_xpath('//div[@class="DdSX2"]/a[contains(@class,"coreSpriteRightPaginationArrow")]')
        next_pic.click()
        time.sleep(2)   
```

```python
unlike_post('dilsefoodie')
```

## 6 :  Extract list of followers
#### 1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
#### 2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def followers(string):
    driver.get('https://www.instagram.com/' + string)  
    time.sleep(3)
    
    #the follower button
    l=driver.find_elements_by_class_name('-nal3') 
    
    l[1].click()
    time.sleep(2)
    curr_ht=driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')  
    
    tot_followers=set()
    while True:
        followerlist=driver.find_elements_by_xpath('//a[contains(@class,"FPmhX")]') 
        for i in followerlist:
            print(i.text)
            
            if len(tot_followers)<70:
                
                tot_followers.add(i.text)
            else:
                break
        if len(tot_followers)<70 :
            
            #scrolling
            driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,arguments[0])',curr_ht)
        else:
            break
        new_ht=driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;') 
        curr_ht=new_ht 
        

    print(len(tot_followers))  
    return tot_followers   
```

```python
tot_followers=followers('foodtalkindia')  
```
